```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
sns.set_context('poster',font_scale=3)
sns.countplot(x='sex',data=tips)
plt.show()
```
